```python
from tokenize import String
import pandas as pd
import matplotlib.pyplot as plt
pd.options.mode.chained_assignment = None 
```

Removing Nan's and preparing data for further manipulations

```python
game_sales = pd.read_csv('sales.csv')
game_sales_clean = pd.DataFrame
game_sales_clean=game_sales.dropna(how='any')#Getting rid of Nan lines
game_sales_clean['Genre'].apply(str)
```

Calculating annual sales for every year

```python
game_sales_year = pd.DataFrame
game_sales_year=game_sales_clean.groupby('Year').sum()#Getting yearly sales
del game_sales_year['Rank']
```

In following section of the code we are receieving sorted list of genres. First 5 elements are the top-5 genres which will be analysed further

```python
sales_temp=pd.DataFrame#temporary variable used for different tasks as a buffer
sales_temp=game_sales_clean.groupby('Genre').sum()
sales_temp=sales_temp['Global_Sales'].sort_values(ascending=False)
Genres=list(sales_temp.index.values)
```

Defining a functions that will build graphics

```python
def build_graph(frame, genre, co):
    plt.figure(genre)
    plt.bar(frame.index,frame['Global_Sales'], color=co)
    plt.title(genre)
    plt.grid()
    plt.xlabel("Year")
    plt.ylabel("Global sales, mil. copies")
    plt.plot()
    plt.figure(f"{genre} global")
    plt.bar(frame.index,frame['Global_Sales_%'], color=co)
    plt.title(f"{genre} global")
    plt.grid()
    plt.xlabel("Year")
    plt.ylabel("Part of global sales, %")
    plt.plot()
    plt.show()
```

Extracting data for the first genre and adding this data to a new dataframe which will be later converted to csv

```python
sales_temp=game_sales_clean[game_sales_clean['Genre']==Genres[0]]
sales_temp=sales_temp.groupby('Year').sum()
del sales_temp['Rank']
sales_temp['Global_Sales_%']=sales_temp['Global_Sales']/game_sales_year['Global_Sales']*100
Top_5=pd.DataFrame()#creating an output dataframe that will store all the data for each genre
Top_5.index=sales_temp.index
Top_5[f'Global_sales_{Genres[0]}']=sales_temp['Global_Sales']
Top_5[f'Global_sales_%_{Genres[0]}']=sales_temp['Global_Sales_%']
build_graph(sales_temp, Genres[0], 'm')
```
